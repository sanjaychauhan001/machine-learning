```python
import pandas as pd 
import numpy as np
import scipy.cluster.hierarchy as sch 
import matplotlib.pyplot as plt 
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("C:\\Users\\sanjay\\Documents\\All working files\\data\\Salary_Data.csv")
df = df.values
```

```python
sch.dendrogram(sch.linkage(df,method='average'))
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
labels_ = cluster.fit_predict(df)
```

```python
labels_
```

```python
plt.scatter(df[:,0], df[:,1], c=cluster.labels_, cmap='rainbow')
```
