```python
class LR:
    def __init__(self,lr,epoch):
        self.lr = lr
        self.epoch = epoch
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
     
    def train(self, X, y):
        self.X = np.insert(X.values,0,1,axis=1)
        self.y = y
        weights = np.ones(self.X.shape[1])
        for i in range(self.epoch):
            y_hat = self.sigmoid(np.dot(self.X,weights))
            weights = weights + self.lr*(np.dot((self.y-y_hat),self.X)/self.X.shape[0])
        return weights[1:], weights[0]
```

```python
import numpy as np

class LR:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))



    def fit(self, X, y):
        # Add a bias term to X
        X_bias = np.c_[np.ones((X.shape[0], 1)), X]

        # Initialize weights
        self.weights = np.zeros(X_bias.shape[1])

        for _ in range(self.num_iterations):
            # Calculate the linear combination
            z = np.dot(X_bias, self.weights)
            # Pass through the sigmoid function
            y_hat = self.sigmoid(z)

            # Compute the gradient
            gradient = np.dot(X_bias.T, (y_hat - y)) / len(y)

            # Update weights using gradient descent
            self.weights -= self.learning_rate * gradient

        return self.weights

    def predict(self, X):
        # Add a bias term to X
        X_bias = np.c_[np.ones((X.shape[0], 1)), X]
        # Calculate the linear combination
        z = np.dot(X_bias, self.weights)
        # Pass through the sigmoid function
        y_hat = self.sigmoid(z)
        # Convert probabilities to binary predictions
        predictions = (y_hat >= 0.5).astype(int)
        return predictions
```
