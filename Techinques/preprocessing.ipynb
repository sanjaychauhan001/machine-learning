```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:\\Users\\sanjay\\Documents\\All working files\\data\\machine learning data\\HR_Employee_Attrition.csv")
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

```python
le = LabelEncoder()
y = le.fit_transform(df['Attrition'])
```

```python
df['BusinessTravel'] = df['BusinessTravel'].replace({"Non-Travel":0,"Travel_Rarely":1,"Travel_Frequently":2})
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace({"Single":0,"Married":1,"Divorced":2})
```

```python
df['Over18'].unique()
```

```python
df.drop(columns=['Over18','Attrition'],inplace=True)
```

```python
df['OverTime'] = df['OverTime'].replace({"No":0,'Yes':1})
```

```python
df.info()
```

```python
df[['Department','EducationField','Gender','JobRole']]
```

```python
ohe = OneHotEncoder(drop='first')
temp1 = ohe.fit_transform(df[['Department','EducationField','Gender','JobRole']]).toarray()
```

```python
temp2 = df.drop(columns=['Department','EducationField','Gender','JobRole'])
```

```python
new_df = np.hstack((temp2.values,temp1))
```

```python
new_df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df,y,test_size=0.2)
```

```python
model = LogisticRegression(max_iter=700)
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
scaler = StandardScaler()
new_df = scaler.fit_transform(new_df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df,y,test_size=0.2)
```

```python
model = LogisticRegression(max_iter=700)
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
```
