```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)

n_sample = 1000
class_0_ratio = 0.9
n_class_0 = int(n_sample * class_0_ratio)
n_class_1 = n_sample - n_class_0
```

```python
n_class_0, n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature1':np.random.normal(loc=0, scale=1,size=n_class_0),
    'feature2':np.random.normal(loc=0, scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature1':np.random.normal(loc=2, scale=1,size=n_class_1),
    'feature2':np.random.normal(loc=2, scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority,replace=True,n_samples=len(df_majority),random_state=12)
```

```python
df_minority_upsampled
```

```python
pd.concat([df_majority,df_minority_upsampled])
```
