```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import xgboost as xgb
```

```python
df = pd.read_csv("C:\\Users\\Sanjay\\Documents\\All working files\projects\\mumbai-real-estate-project\\data_for_model.csv")
```

```python
df
```

```python
df = df.drop(columns=['bedrooms'])
```

```python
df
```

```python
X = df.drop(columns=['price'],axis=1)
y = df['price']
```

```python
X.head(10)
```

```python
ordinal = OrdinalEncoder()
X['flat_type'] = ordinal.fit_transform(X['flat_type'].values.reshape(-1,1))
```

```python
X['which_floor'] = ordinal.fit_transform(X['which_floor'].values.reshape(-1,1))
```

```python
X['furnishing'] = X['furnishing'].map({'Unfurnished':0,'Semi Furnished':1,'Fully Furnished':2})
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
ohe.fit_transform(X['address'].values.reshape(-1,1))
```

```python
pd.get_dummies(X['address'])
```

```python
X = pd.merge(X,pd.get_dummies(X['address']),left_index=True,right_index=True).drop(columns=['address'])
```

```python
lr = LinearRegression()
k = KFold(n_splits=10,shuffle=False)
np.mean(cross_val_score(lr,X,y,scoring='r2',cv=k))
```

```python
lr = Ridge()
k = KFold(n_splits=10,shuffle=True)
param = {'alpha':[0.001,0.01,0.3,2,4]}
g = GridSearchCV(estimator=lr,param_grid=param,scoring='r2',cv=k)
np.mean(cross_val_score(g,X,y,scoring='r2',cv=k))
```

```python
lr = KNeighborsRegressor()
k = KFold(n_splits=10,shuffle=True)
param = {'n_neighbors':[3,5,7,9],
        'weights':["uniform", "distance"]}
g = GridSearchCV(estimator=lr,param_grid=param,scoring='r2',cv=k)
print(np.mean(cross_val_score(g,X,y,scoring='r2',cv=k)))
```

```python
lr = DecisionTreeRegressor()
k = KFold(n_splits=10,shuffle=True)
param = {'criterion':["gini", "entropy",],
        'splitter':["best", "random"],
        "max_depth":[3,5,7,9,10,None],
        "min_samples_split":[3,5,7,9],
        "min_samples_leaf":[1,2,3,4,5,6],
        'max_features':["sqrt","log2"],
        "max_leaf_nodes":[5,7,9,10]}
g = GridSearchCV(estimator=lr,param_grid=param,scoring='r2',cv=k)
print(np.mean(cross_val_score(g,X,y,scoring='r2',cv=k)))
```

```python
lr = RandomForestRegressor()
k = KFold(n_splits=10,shuffle=True)
param = {'criterion':["gini", "entropy", "log_loss"],
         "n_estimators":[30,50,100,150,200],
        'splitter':["best", "random"],
        "max_depth":[3,5,7,9,10,None],
        "min_samples_split":[3,5,7,9],
        "min_samples_leaf":[1,2,3,4,5,6],
        'max_features':["sqrt","log2"],
        "max_leaf_nodes":[5,7,9,10]}
g = GridSearchCV(estimator=lr,param_grid=param,scoring='r2',cv=k)
print(np.mean(cross_val_score(g,X,y,scoring='r2',cv=k)))
```
