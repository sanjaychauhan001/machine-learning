```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)

plt.scatter(X[:,0], X[:,1], c=y, cmap='winter')
```

```python
from sklearn.svm import SVC

s = SVC(kernel='linear', C=1)
s.fit(X,y)
```
