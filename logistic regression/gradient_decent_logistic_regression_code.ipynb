```python
def gd(X,y):
    X = np.insert(X,0,1,axis=1)
    weight = np.ones(X.shape[1])
    lr = 0.5

    for i in range(2500):
        y_hat = sigmoid(np.dot(X, weight))

        weight = weight + lr * (np.dot((y - y_hat),X))/X.shape[0]

    return weight[1:], weight[0]    
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
coef_, intercept = gd(X,y)
```
