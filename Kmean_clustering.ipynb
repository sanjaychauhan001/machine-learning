```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# parameters
n_sample = 300
n_features = 2
centers = 6

# generate data

X,y = make_blobs(n_samples=n_sample, n_features=n_features, centers=centers, cluster_std=1, random_state=2)

# visualize data
plt.scatter(X[:,0], X[:,1], s=50, cmap = 'rainbow')
plt.show()
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=4, n_init=10)
y_mean = kmeans.fit_predict(X)
```

```python
y_mean
```

```python
plt.scatter(X[:,0], X[:,1], c=y_mean, cmap="rainbow")
```

```python
# parameters
n_sample = 300
n_features = 2
centers = 4

# generate data

X,y = make_blobs(n_samples=n_sample, n_features=n_features, centers=centers, cluster_std=1, random_state=4)

# visualize data
plt.scatter(X[:,0], X[:,1], s=50, cmap = 'rainbow')
plt.show()
```

```python
ss = []

for i in range(1,11):
    kmeans = KMeans(n_init=10,n_clusters=i)
    kmeans.fit(X)
    ss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), ss)
```

```python
sil = []

for i in range(2,11):
    kmean = KMeans(n_clusters=i)
    cluster_label = kmean.fit_predict(X)
    avg_sil = silhouette_score(X,cluster_label)
    sil.append(avg_sil)
```

```python
plt.plot(range(2,11), sil)
```
