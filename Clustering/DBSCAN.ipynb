```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_circles
from sklearn.cluster import KMeans, DBSCAN
```

```python
X,y = make_blobs(n_samples=200, n_features=2, centers=2, cluster_std=1.2,random_state=2)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
k = KMeans(n_clusters=2,n_init=10)
y_mean = k.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_mean, cmap='rainbow')
```

```python
d = DBSCAN(eps=2, min_samples=8)
y_mean = d.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_mean, cmap='rainbow')
plt.show()
```

```python
X,y = make_circles(n_samples=500, factor=0.5, noise=0.04)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
k = KMeans(n_clusters=2,n_init=10)
y_mean = k.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_mean, cmap='rainbow')
```

```python
d = DBSCAN(eps=0.3, min_samples=5)
y_mean1 = d.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_mean1, cmap='rainbow')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(X,y_mean))
print(silhouette_score(X,y_mean1))
```
